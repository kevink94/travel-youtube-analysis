```python
# function to get channel statistics

def get_channel_stats(youtube, channel_ids):
    
    all_data=[]
    
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=','.join(channel_ids)
    )
    response = request.execute()
    #JSON(response)
    
    #loop through items
    for item in response['items']:
        data={ 'channelname': item['snippet']['title'],
               'subscribers': item['statistics']['subscriberCount'],
               'views': item['statistics']['viewCount'],
               'totalvideos': item['statistics']['videoCount'], 
               'playlistId': item['contentDetails']['relatedPlaylists']['uploads']
             }
        
        all_data.append(data)
    
    return pd.DataFrame(all_data)
```

### Get video statistics for all the channels

```python
# function to acquire video ids using the playlist ids

def get_video_ids(youtube, playlist_id):
    
    video_ids=[]
    
    request = youtube.playlistItems().list(
        part="snippet,contentDetails",
        playlistId= playlist_id,
        maxResults = 50
    )
    response = request.execute()
    #JSON(response)
    
    for item in response['items']:
        video_ids.append(item['contentDetails']['videoId'])
    
    next_page_token = response.get('nextPageToken')
    while next_page_token is not None:
        request = youtube.playlistItems().list(
            part="snippet,contentDetails",
            playlistId = playlist_id,
            maxResults = 50,
            pageToken = next_page_token
        )
        response = request.execute()
        #JSON(response)
    
        for item in response['items']:
            video_ids.append(item['contentDetails']['videoId'])
        
        next_page_token = response.get('nextPageToken')
    
    return video_ids
```

```python
# getting video ids for each travel blogger into a list

playlist_ids=channel_df['playlistId'].tolist()
channel_video_ids={}

for playlist_id in playlist_ids:
    video_ids= get_video_ids(youtube, playlist_id)
    name=channel_df[channel_df['playlistId']==playlist_id]['channelname'].tolist()[0]
    channel_video_ids[name]= video_ids
```

```python
# getting video information such as video titles, likes count, comment counts etc.

def get_video_details(youtube, video_ids):
    
    all_video_info=[]
        
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id= ','.join(video_ids[i:i+50])
        )
        response = request.execute()

   

        #JSON(response)
        for video in response['items']:
            stats_to_keep={ 'snippet' : ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                            'statistics' : ['viewCount', 'likeCount', 'favouriteCount', 'commentCount'],
                            'contentDetails' : ['duration', 'definition', 'caption']
                          }
            video_info = {}
            video_info['video_id']= video['id']
    
            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v]=video[k][v]
                    except:
                        video_info[v]= None
        
            all_video_info.append(video_info)
        
    return pd.DataFrame(all_video_info)
```

```python
# creating a videos dataframe with all the data

video_ids=[]

for v in channel_video_ids.values():
    video_ids+=v
#len(video_ids)
videos_df= get_video_details(youtube, video_ids)
videos_df
```

Let's get the comments of each video too into `comment_df` as well.

```python
# Get each videos comments

def get_comments_in_videos(youtube, video_ids):
    all_comments=[]
    
    for video_id in video_ids:
        request = youtube.commentThreads().list(
            part="snippet, replies",
            videoId= video_id
        )
        ''' Some videos may contain disabled comments therefore we have to use a 
        try/ except statement to filter disablesd comments videos '''
        try:
            response = request.execute()
        except:
            response = None
        
        if response is not None:
            comments_in_vid = [comment['snippet']['topLevelComment']['snippet']['textOriginal'] for comment in response['items']]
        else:
            comments_in_vid = 'disabled'
        comments_in_vid_info = {'video_id': video_id, 'comments': comments_in_vid}
        
        all_comments.append(comments_in_vid_info)
    
    return pd.DataFrame(all_comments)
```

```python
comments_df = get_comments_in_videos(youtube, video_ids)
comments_df
```

## Enriching the Data

I want to enrich the data for further analyses, for example:

- create published date column with another column showing the day in the week the video was published, which will be useful for later analysis.

- convert video duration to seconds instead of the current default string format

- calculate number of tags for each video

- calculate comments and likes per 1000 view ratio

- calculate title character length

We need to check if the data type of the columns are correct. I have checked the data types and indeed some count columns such as view count and comment count are currently not in correct data type. In this step, we convert these count columns into integer.

```python
numeric_cols=['viewCount', 'likeCount', 'favouriteCount', 'commentCount']
videos_df[numeric_cols]=videos_df[numeric_cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
# creating a new column for published day of the week
videos_df['publishedAt'] = videos_df['publishedAt'].apply(lambda x: parser.parse(x))
videos_df['publishDayName'] = videos_df['publishedAt'].apply(lambda x: x.strftime('%A'))
```

```python
# Creating a new column with duration in seconds
import isodate
videos_df['durationSecs'] = videos_df['duration'].apply(lambda x: isodate.parse_duration(x))
videos_df['durationSecs'] = videos_df['durationSecs'].astype('timedelta64[s]')
```

```python
# Add number of tags
videos_df['tagsCount'] = videos_df['tags'].apply(lambda x: 0 if x is None else len(x))
```

```python
# Comments and likes per 1000 view ratio
videos_df['likeRatio'] = videos_df['likeCount']/ videos_df['viewCount'] * 1000
videos_df['commentRatio'] = videos_df['commentCount']/ videos_df['viewCount'] * 1000
```

```python
# Title character length
videos_df['titleLength'] = videos_df['title'].apply(lambda x: len(x))
```

### Wordcloud for video titles

As I'm interested to see what the creators are making videos about and which terms most frequently appear in their video titles, I will create a wordcloud for the most common words. We first need to remove the stopwords which do note contribute a lot to the meaning of the title.

```python
stop_words = set(stopwords.words('english'))
# we add 'iam_marwa' to the stop_words to be filtered since he uses his name in every title
stop_words.add('iam_marwa')
videos_df['title_no_stopwords'] = videos_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in videos_df['title_no_stopwords'].tolist() for a in b])
all_words_str= ''.join(all_words)

def plot_cloud(wordcloud):
    plt.figure(figsize=(30,20))
    plt.imshow(wordcloud)
    plt.axis('off');

wordcloud = WordCloud(width= 2000, height= 1000, random_state= 1, background_color= 'black',
                      colormap= 'viridis', collocations= False).generate(all_words_str)
plot_cloud(wordcloud)
```

From above we see Sri Lanka, Kenya, India, etc and among others are the most used words in titles by these travel bloggers showing that they are their most frequented travel destinations
